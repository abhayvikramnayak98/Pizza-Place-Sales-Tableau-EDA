```python
data_folder_path = r'pizza_sales'
df1 = pd.read_csv(os.path.join(data_folder_path, 'order_details.csv'))
df2 = pd.read_csv(os.path.join(data_folder_path, 'orders.csv'))
df3 = pd.read_csv(os.path.join(data_folder_path, 'pizza_types.csv'), encoding='latin1')
df4 = pd.read_csv(os.path.join(data_folder_path, 'pizzas.csv'))
```

```python
df5 = pd.merge(df4, df3, on='pizza_type_id', how='inner')
df5.head()
```

```python
df6 = pd.merge(df1, df5, on='pizza_id', how='inner').sort_values(by = 'order_details_id').reset_index(drop = True)
df6.head()
```

```python
merged_df = pd.merge(df6, df2, on='order_id', how='inner').sort_values(by = 'order_details_id').reset_index(drop = True)
merged_df.head()
```

```python
merged_df['total_price'] = merged_df['quantity'] * merged_df['price']
merged_df.head()
```

```python
merged_df.columns
```

```python
merged_df.dtypes
```

```python
desired_columns = ['order_details_id', 'order_id', 'pizza_id', 'quantity', 'size', 'price', 'total_price', 'name', 'pizza_type_id', 'category', 'ingredients', 'date', 'time']
merged_df = merged_df.reindex(columns = desired_columns)
merged_df.head()
```

```python
merged_df['order_time'] = pd.to_datetime(merged_df['date'] + ' ' + merged_df['time'])
merged_df.drop(['date', 'time'], axis=1, inplace=True)
merged_df.head()
```

```python
merged_df['size'] = merged_df['size'].replace({'L': 'Large', 'M': 'Medium', 'S': 'Small', 'XL': 'X-Large', 'XXL': 'XX-Large'})
merged_df.head()
```
